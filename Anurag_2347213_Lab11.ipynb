```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("health_fitness_dataset.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Wellness Score'], marker='o', linestyle='-', color='b')
plt.xlabel('UserID')
plt.ylabel('Wellness Score')
plt.title('Wellness Score of All Users')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_wellness_by_age = df.groupby('Age')['Wellness Score'].mean()
average_wellness_by_age.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Age')
plt.ylabel('Average Wellness Score')
plt.title('Average Wellness Score by Age Group')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df[df['Gender'] == 'Male']['Weight (kg)'],
            df[df['Gender'] == 'Male']['Height (cm)'],
            label='Male', c='blue', alpha=0.5)

plt.scatter(df[df['Gender'] == 'Female']['Weight (kg)'],
            df[df['Gender'] == 'Female']['Height (cm)'],
            label='Female', c='red', alpha=0.5)

plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Scatter Plot of Weight vs. Height by Gender')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
numeric_columns = ['Age', 'BMI', 'Wellness Score']
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
